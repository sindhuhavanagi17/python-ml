#Numpy


#Create Arrays

```python
import numpy as np

a=np.array([1,2,3,4,5])
b=np.arange(1,6)
c=np.zeros((2,3))
d=np.ones((3,2))

print("Array a:",a)
print("Array b:",b)
print("2x3 zeros:",c)
print("3x2 ones:",d)
```

#Indexing & Slicing

```python
print("First element of a:",a[0])
print("Last two elements of a:",a[-2])
```

#Basic arthimetic operations

```python
print("a+2:",a+2)
print("a*3:",a*3)
```

#Basic statistics

```python
print("square of a:", np.square(a))
print("mean of a:", np.mean(a))
print("standard deviation:", np.std(a))
```

#Reshaping arrays

```python
matrix=np.arange(1,10).reshape((3,3))
print("Reshaped 3x3 martix:")
matrix
```

#Transpose

```python
print("Transpose:")
matrix.T
```

#Boolean indexing

```python
print("Elements>5:",matrix[matrix>5])
```

#Create a DataFrame manually

```python
import pandas as pd

data={
    'Name':['Alice','Bob','Charlie','David','Eva'],
    'Age':[25,30,35,40,22],
    'Department':['teacher','Engineering','HR','Engineering','sales'],
    'Salary':[5000,8000,4000,9000,45000]
}
df=pd.DataFrame(data)
df
```

#Basic operations

```python
print("First 3 rows:")
df.head(3)
```

```python
print("columns names:")
df.columns
```

```python
print("Data types:")
df.dtypes
```

```python
print("Describe numerical columns:")
df.describe()
```

#Filtering

```python
print("People with salary > 50000:")
df[df['Salary']>50000]
```

#Sorting

```python
print("Sorted by Age:")
df.sort_values(by='Age')
```

#Grouping and Aggregation

```python
grouped=df.groupby('Department')['Salary'].mean()
print("Average Salary by Department:")
grouped
```

#Adding new columns

```python
df['Salary in lakhs']=df['Salary']/10000
df
```

#Line Plot

```python
import matplotlib.pyplot as plt

years=[2019,2020,2021,2022]
revenue=[100,120,130,160]

plt.plot(years,revenue,marker='o',linestyle='--',color='purple')
plt.title("Company Revenue Over Years")
plt.xlabel("Year")
plt.ylabel("Revenue(in crores)")
plt.xticks(years)
plt.grid(True)
plt.show()
```

#Bar Chart

```python
departments=df['Department'].value_counts()
departments.plot(kind='bar',color='pink')
plt.title("Employeem Count by Department")
plt.xlabel("Department")
plt.ylabel("Count")
plt.show()
```

#Histogram

```python
ages=df['Age']
plt.hist(ages,bins=5,color='blue',edgecolor='black')
plt.title("Age Distrubtion")
plt.xlabel("Age")
plt.ylabel("Frequency")
plt.show()
```

Pie chart

```python
df.groupby('Department').size().plot(kind='pie',autopct='%1.1f%%')
plt.title("Department Proportion")
plt.ylabel("")
plt.show()
```

#Load sample dataset

```python
import seaborn as sns
tips=sns.load_dataset("tips")
tips.head()
```

```python
print("\nInfo about the tips dataset:")
tips.info()
print("\nDescription of the tips dataset:")
tips.describe(include='all')
```

```python
sns.pairplot(tips,hue="sex")
plt.show()
```

```python
corr=tips.corr(numeric_only=True)
sns.heatmap(corr,annot=True,cmap="coolwarm")
plt.title("correlation matrix")
plt.show()
```

#Countplot

```python
sns.countplot(x='day',data=tips,palette='Set2',hue='day',legend=False)
plt.title("Count of Records by Day")
plt.show()
```

#Boxplot

```python
sns.boxplot(x='day',y='total_bill',data=tips)
plt.title("Boxplot of Total Bill by Day")
plt.show()
```
